```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import pandas as pd

st = pd.read_csv ("/content/healthcare-dataset-stroke-data.csv")
st.head()
```

```python
st.isnull().sum()
```

```python
st.dropna(inplace= True)
st.isnull().sum()
```

```python
le = LabelEncoder()
st['hypertension'] = le.fit_transform(st['hypertension'])
st['gender']=le.fit_transform(st['gender'])
st['heart_disease']=le.fit_transform(st['heart_disease'])
st['ever_married']=le.fit_transform(st['ever_married'])
st['work_type']=le.fit_transform(st['work_type'])
st['Residence_type']=le.fit_transform(st['Residence_type'])
st['avg_glucose_level']=le.fit_transform(st['avg_glucose_level'])
st['bmi']=le.fit_transform(st['bmi'])
st['smoking_status']=le.fit_transform(st['smoking_status'])
st.head()
```

```python
ind = st[['id','gender','age','hypertension',	'heart_disease','ever_married','work_type','Residence_type','avg_glucose_level','bmi','smoking_status']]
dep = st['stroke']
LR = LinearRegression()
LR.fit(ind,dep)
LR.predict([[32,0,0,2,1,0,34,67,2,4,6]])
```
